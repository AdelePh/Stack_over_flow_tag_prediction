# Pre processing text

```python
# Pre-processing text - Clean text 
def extract_entities(doc):
    return [e.text for e in doc.ents if e.label_ in ("PER", "LOC", "DATE", "ORG", "NORP", "GPE", "MONEY", "ORDINAL","PERCENT")]

def clean_html_tags(text):
    soup = BeautifulSoup(text, features="html.parser")
    return soup.get_text().lower()

def pre_processing(doc,en_nlp_model,en_sw_spacy,en_punc,en_stemmer,labels):
    entities_to_remove = extract_entities(doc)
    tokens=[]
    for token in doc:
        if token.text in labels:
            tokens.append(token.text)
        else:
            if (token.is_alpha and len(token.text)>1
            and not token.is_punct and not token.is_currency
            and not token.is_digit and not token.like_email
            and not token.like_url and not token.is_space
            and not token.is_stop and not token.like_num):
                tokens.append(token.lemma_)
    clean_text = " ".join(en_stemmer.stem(token) for token in tokens if token not in entities_to_remove)
    return clean_text
```

```python
'''
Apply clean text functions  
'''
en_nlp_model = spacy.load('en_core_web_sm') # load nlp model
#custom tokenizer from en_core_web_sm so that c#, .net, objective-c and c++ are not tokenized, they are kept as 1 token
en_nlp_model.tokenizer.add_special_case("c#", [{ORTH: "c#"}]) 
en_nlp_model.tokenizer.add_special_case(".net", [{ORTH: ".net"}]) 
en_nlp_model.tokenizer.add_special_case("objective-c", [{ORTH: "objective-c"}]) 
en_nlp_model.tokenizer.add_special_case("c++", [{ORTH: "c++"}]) 
en_sw_spacy = en_nlp_model.Defaults.stop_words # stop words
en_punc = string.punctuation.split(',') # punctuations
# Initialize the stemmer
en_stemmer = PorterStemmer()
merged_df["Body"] = merged_df["Body"].str.lower() # convert text in column 'Body' to lowercase
merged_df["Title"] = merged_df["Title"].str.lower() # convert text in column 'Title' to lowercase
body_docs= list(merged_df["Body"])
title_docs = list(merged_df["Title"])
#remove html tags
merged_df["Body"] = merged_df["Body"].apply(clean_html_tags) 
merged_df["Title"] = merged_df["Title"].apply(clean_html_tags)
clean_body_text = [] #saves text in 'Body' col after cleaning
clean_title_text = [] #saves text in 'Title' col after cleaning
#clean text in 2 cols "Body" and "Title" by appling pre_processing to the whole col
n_process = multiprocessing.cpu_count()
for doc in en_nlp_model.pipe(body_docs,n_process=n_process):
    clean_body_text.append(pre_processing(doc,en_nlp_model, en_sw_spacy, en_punc, en_stemmer, keep_tags))
for doc in en_nlp_model.pipe(title_docs,n_process=n_process):
    clean_title_text.append(pre_processing(doc,en_nlp_model, en_sw_spacy, en_punc, en_stemmer,keep_tags))
merged_df["Title"] = clean_title_text
merged_df["Body"] = clean_body_text
```

```python
# drop unnecessary columns
merged_df.drop(columns=["CreationDate","Score","ClosedDate","OwnerUserId"], inplace=True)
```

# Construct Machine Learning Engines

```python
# Encode categorical features using One Hot Encoding in Pandas
encoded_df_dummies = pd.get_dummies(merged_df, columns=["Tag"])
```

```python
columns = list(encoded_df_dummies.columns)#list of columns 
topic_columns = columns[3:] # topics

X1 = encoded_df_dummies.groupby('Id').agg({'Title': 'first'})["Title"]
X2 = encoded_df_dummies.groupby('Id').agg({'Body': 'first'})["Body"]
y = np.asarray(encoded_df_dummies.groupby('Id')[topic_columns].sum())

# Split data into train and test sets
X1_train, X1_test, X2_train, X2_test, y_train, y_test = train_test_split(X1, X2, y, test_size = 0.2, random_state = 0)
# Initialize TfidfVectorizer object
vectorizer_X1 = TfidfVectorizer(ngram_range=(1, 2))
vectorizer_X2 = TfidfVectorizer(ngram_range=(1, 2))
# Fit the vectorizer to the training data and transform it into a matrix
X1_train_tfidf = vectorizer_X1.fit_transform(X1_train)
X2_train_tfidf = vectorizer_X2.fit_transform(X2_train)
# feature selection using chi square and SelectKBest
selector_X1 = SelectKBest(chi2, k=6000)
X1_train_selected = selector_X1.fit_transform(X1_train_tfidf, y_train)
selector_X2 = SelectKBest(chi2, k=6000)
X2_train_selected = selector_X2.fit_transform(X2_train_tfidf,y_train)

# Transform the test data into a matrix
X1_test_tfidf = vectorizer_X1.transform(X1_test)
X2_test_tfidf = vectorizer_X2.transform(X2_test)

# Apply feature selection to test data
X1_test_selected = selector_X1.transform(X1_test_tfidf)
X2_test_selected = selector_X2.transform(X2_test_tfidf)

# combine X1 and X2 into single arrays
X_train_selected = hstack([X1_train_selected,X2_train_selected])
X_test_selected = hstack([X1_test_selected,X2_test_selected])
```

```python
# performance of some classification models 
models = [
          ('LogisticRegression', LogisticRegression(penalty="l2")), 
          ('MNB', MultinomialNB(alpha=0.007)),
          ('KNN', KNeighborsClassifier()),
          ('SVC', SVC()), 
          ('RF', RandomForestClassifier(n_estimators=100, min_samples_split=5, min_samples_leaf=2)),
          ('GNB', GaussianNB())
        ]
export_models = [] 
for name, model in models:
    clf = LabelPowerset(model).fit(X_train_selected, y_train)
    predictions_train = clf.predict(X_train_selected)
    predictions = clf.predict(X_test_selected)
    pred_score = precision_score(y_test,predictions,average = 'samples')
    print(name)
    print("Precision Score in training set = ",precision_score(y_train,predictions_train,average = 'samples'))
    print("Precision Score in test set = ",pred_score)
    print('------------------------------------------------------------')
    export_models.append(clf)
```

```python
# overcome overfitting in MultinomialNB model by using k-fold cross validation
# feature selection using chi square and SelectKBest
selector_X1 = SelectKBest(chi2, k=2600)
X1_train_selected = selector_X1.fit_transform(X1_train_tfidf, y_train)
selector_X2 = SelectKBest(chi2, k=2600)
X2_train_selected = selector_X2.fit_transform(X2_train_tfidf,y_train)

# Transform the test data into a matrix
X1_test_tfidf = vectorizer_X1.transform(X1_test)
X2_test_tfidf = vectorizer_X2.transform(X2_test)

# Apply feature selection to test data
X1_test_selected = selector_X1.transform(X1_test_tfidf)
X2_test_selected = selector_X2.transform(X2_test_tfidf)

# combine X1 and X2 into single arrays
X_train_selected = hstack([X1_train_selected,X2_train_selected])
X_test_selected = hstack([X1_test_selected,X2_test_selected])
X = vstack((X_train_selected, X_test_selected)).toarray()
y = vstack((y_train, y_test)).toarray()
# evaluate a logistic regression model using k-fold cross-validation
from sklearn.metrics import precision_recall_fscore_support
# Define the k-fold cross-validation split
n_folds = 70
kfold = KFold(n_splits=n_folds, shuffle=True)
#skf = StratifiedKFold(n_splits=3)
# Initialize a list to store the evaluation metrics for each fold
metrics = []
clf = LabelPowerset(MultinomialNB(alpha=0.007)).fit(X_train_selected, y_train)

# Iterate over the folds
for train_index, val_index in kfold.split(X,y):
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = y[train_index], y[val_index]
    
    # Fit the classifier on the training data
    clf.fit(X_train, y_train)
    
    # Make predictions on the validation data
    y_pred = clf.predict(X_val)
    
    # Compute the evaluation metrics
    precision, recall, fscore, _ = precision_recall_fscore_support(y_val, y_pred, average='micro')
    
    # Store the evaluation metrics for this fold
    metrics.append((precision, recall, fscore))
    
# Compute the average evaluation metrics over all folds
precision = sum(x[0] for x in metrics)/n_folds
recall = sum(x[1] for x in metrics) /n_folds
fscore = sum(x[2] for x in metrics) /n_folds
print("precision score: ", precision)
print("recall score: ", recall)
print("fscore: ", fscore)
```
